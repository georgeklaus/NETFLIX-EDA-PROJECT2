```python
#Q(4a)
# Number of Movies and TV Shows
content_type_counts = df['type'].value_counts()
content_type_counts
```

```python
#Q4(b)
#Number of contents by country
country_counts = df['country'].value_counts().head(10)  # Top 10 countries
country_counts
```

```python
#Q4(c)
#number of contents released each year
release_year_counts = df['release_year'].value_counts().sort_index()
release_year_counts
```

```python
#Q4(d)
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

# Handling NaN values in 'duration' column for movies
# Replace NaN with a placeholder value before converting
movies.loc[:, 'duration'] = movies['duration'].str.replace(' min', '')
movies.loc[:, 'duration'] = movies['duration'].fillna('0').astype(int)

# Ploting the duration for movies and year
movies_duration_trend = movies.groupby('release_year')['duration'].mean()
movies_duration_trend.plot(kind='line', title='Average Movie Duration Over Years')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (min)')
plt.show()
```

```python
#Q4(e)
# Split the 'listed_in' column to get individual genres
# Top 10 genres
all_genres = df['listed_in'].str.split(', ').explode()
genre_counts = all_genres.value_counts().head(10) 
print(genre_counts)
```

```python
#Q4(f)
# Count genres over the years
genres_over_years = df[['release_year', 'listed_in']].dropna()
genres_over_years['listed_in'] = genres_over_years['listed_in'].str.split(', ')

# Getting one genre per row throgh Explode
genres_over_years = genres_over_years.explode('listed_in')

# Count of genres by year
genre_trends = genres_over_years.groupby(['release_year', 'listed_in']).size().unstack().fillna(0)
genre_trends = genre_trends.loc[:, genre_trends.sum().sort_values(ascending=False).head(10).index]  # Top 10 genres

# Plot trends
genre_trends.plot(kind='line', figsize=(12, 8), title='Trends in Popularity of Specific Genres Over Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python
#Q4(g)
# Distribution of content by country
country_preference = df['country'].value_counts().head(10)
country_preference.plot(kind='bar', title='Top 10 countries Contributing Content')
plt.xlabel('country')
plt.ylabel('Number of Titles')
plt.show()
```
